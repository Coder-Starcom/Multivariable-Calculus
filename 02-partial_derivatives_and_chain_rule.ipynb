# Multivariable Chain Rule


# 1. Define the Functions and Variables

We have an outer function $f$ that depends on two intermediate variables, $x$ and $y$, and these variables, in turn, depend on the single parameter $t$.

*   **Outer Function:** $f(x, y) = x \cdot y$
*   **Intermediate Variables:**
    $$
    x(t) = e^{-t/5} \cos(\pi t)
    $$
    $$
    y(t) = e^{-t/5} \sin(\pi t)
    $$

---

# 2. The Chain Rule Formula

The rate of change of $f$ with respect to $t$ is given by:

$$
\frac{df}{dt} = \frac{\partial f}{\partial x} \frac{dx}{dt} + \frac{\partial f}{\partial y} \frac{dy}{dt}
$$

---

# 3. Calculate the Partial Derivatives of $f$

We need $\frac{\partial f}{\partial x}$ and $\frac{\partial f}{\partial y}$ from $f(x, y) = x y$.

1.  **Partial derivative with respect to $x$:** Treat $y$ as a constant.
    $$
    \frac{\partial f}{\partial x} = \frac{\partial}{\partial x} (xy) = y
    $$

2.  **Partial derivative with respect to $y$:** Treat $x$ as a constant.
    $$
    \frac{\partial f}{\partial y} = \frac{\partial}{\partial y} (xy) = x
    $$

---

# 4. Calculate the Ordinary Derivatives of $x$ and $y$

We need $\frac{dx}{dt}$ and $\frac{dy}{dt}$. We must use the **product rule** for both, since $x(t)$ and $y(t)$ are products of $e^{-t/5}$ and a trigonometric function.

### Derivative of $x(t)$

$$
x(t) = \underbrace{e^{-t/5}}_{u} \underbrace{\cos(\pi t)}_{v}
$$

Using the product rule $\frac{d}{dt}(uv) = u'v + uv'$:

$$
\frac{dx}{dt} = \left( -\frac{1}{5} e^{-t/5} \right) \cos(\pi t) + e^{-t/5} \left( -\pi \sin(\pi t) \right)
$$

$$
\frac{dx}{dt} = e^{-t/5} \left[ -\frac{1}{5} \cos(\pi t) - \pi \sin(\pi t) \right]
$$

### Derivative of $y(t)$

$$
y(t) = \underbrace{e^{-t/5}}_{u} \underbrace{\sin(\pi t)}_{v}
$$

Using the product rule $\frac{d}{dt}(uv) = u'v + uv'$:

$$
\frac{dy}{dt} = \left( -\frac{1}{5} e^{-t/5} \right) \sin(\pi t) + e^{-t/5} \left( \pi \cos(\pi t) \right)
$$

$$
\frac{dy}{dt} = e^{-t/5} \left[ -\frac{1}{5} \sin(\pi t) + \pi \cos(\pi t) \right]
$$

---

# 5. Substitute into the Chain Rule Formula

Now we substitute the results from Steps 3 and 4 into the chain rule formula:

$$
\frac{df}{dt} = \underbrace{y}_{\frac{\partial f}{\partial x}} \cdot \underbrace{\frac{dx}{dt}}_{\text{from Step 4}} + \underbrace{x}_{\frac{\partial f}{\partial y}} \cdot \underbrace{\frac{dy}{dt}}_{\text{from Step 4}}
$$

Substitute $x$ and $y$ back in terms of $t$:

$$
\frac{df}{dt} = \left( e^{-t/5} \sin(\pi t) \right) \cdot \left( e^{-t/5} \left[ -\frac{1}{5} \cos(\pi t) - \pi \sin(\pi t) \right] \right) + \left( e^{-t/5} \cos(\pi t) \right) \cdot \left( e^{-t/5} \left[ -\frac{1}{5} \sin(\pi t) + \pi \cos(\pi t) \right] \right)
$$

### Simplify the Expression

Factor out the common term $e^{-t/5} \cdot e^{-t/5} = e^{-2t/5}$:

$$
\frac{df}{dt} = e^{-2t/5} \left[ \sin(\pi t) \left( -\frac{1}{5} \cos(\pi t) - \pi \sin(\pi t) \right) + \cos(\pi t) \left( -\frac{1}{5} \sin(\pi t) + \pi \cos(\pi t) \right) \right]
$$

Distribute the terms inside the brackets:

$$
\frac{df}{dt} = e^{-2t/5} \left[ -\frac{1}{5} \sin(\pi t) \cos(\pi t) - \pi \sin^2(\pi t) - \frac{1}{5} \cos(\pi t) \sin(\pi t) + \pi \cos^2(\pi t) \right]
$$

Combine the like terms:

$$
\frac{df}{dt} = e^{-2t/5} \left[ \pi (\cos^2(\pi t) - \sin^2(\pi t)) - \frac{2}{5} \sin(\pi t) \cos(\pi t) \right]
$$

Using the double-angle identities ($\cos(2\theta) = \cos^2(\theta) - \sin^2(\theta)$ and $\sin(2\theta) = 2 \sin(\theta) \cos(\theta)$):

$$
\frac{df}{dt} = e^{-2t/5} \left[ \pi \cos(2\pi t) - \frac{1}{5} (2 \sin(\pi t) \cos(\pi t)) \right]
$$

$$
\frac{df}{dt} = e^{-2t/5} \left[ \pi \cos(2\pi t) - \frac{1}{5} \sin(2\pi t) \right]
$$

## Plot v(t)

```python
t = np.arange(0, 5, 0.025)
V = [np.exp(-t / 5) * np.cos(np.pi * t), np.exp(-t / 5) * np.sin(np.pi * t)]

plt.plot(V[0], V[1])
for v_x, v_y, t_i in zip(V[0][::20], V[1][::20], t[::20]):
    plt.text(v_x, v_y, 't={}'.format(t_i))
plt.show()
```

## Plot f(v(t)) 

```python
F = V[0] * V[1]

plt.plot(t, F)
plt.xlabel('t')
plt.ylabel('f(v(t))')
plt.show()
```

## Derivative of f(v(t)) wrt t

```python
df = -np.exp(-2 * t / 5.0) * \
     (2 / 5.0 * np.cos(np.pi * t) * np.sin(np.pi * t) - \
     np.pi * np.cos(2 * np.pi * t))

plt.plot(t, df)
plt.xlabel('t')
plt.ylabel('df(v(t))/dt')
plt.show()
```


# What is a Vector Field?

First, let's quickly remember what a vector field is. A vector field, often denoted as $\mathbf{F}(x, y, z)$, is a function that assigns a **vector** to every point in space. 

*   **Example:** A wind map where at every location $(x, y)$, there's an arrow showing the speed and direction of the wind. 

A 2D vector field $\mathbf{F}$ looks like this:

$$
\mathbf{F}(x, y) = P(x, y) \mathbf{i} + Q(x, y) \mathbf{j} = \langle P(x, y), Q(x, y) \rangle
$$

Here, $P(x, y)$ and $Q(x, y)$ are the scalar functions that define the $x$ and $y$ components of the vector at any point.

---

# Partial Derivative of a Vector Field

Taking the partial derivative of a vector field is surprisingly simple: **You just take the partial derivative of each component function separately.** 

If we want to find the partial derivative of $\mathbf{F}(x, y)$ with respect to $x$, denoted $\frac{\partial \mathbf{F}}{\partial x}$, we do this:

$$
\frac{\partial \mathbf{F}}{\partial x} = \frac{\partial}{\partial x} \langle P(x, y), Q(x, y) \rangle = \left\langle \frac{\partial P}{\partial x}, \frac{\partial Q}{\partial x} \right\rangle
$$

Similarly, for the partial derivative with respect to $y$:

$$
\frac{\partial \mathbf{F}}{\partial y} = \frac{\partial}{\partial y} \langle P(x, y), Q(x, y) \rangle = \left\langle \frac{\partial P}{\partial y}, \frac{\partial Q}{\partial y} \right\rangle
$$

### What Does It Mean? 

The resulting vector, $\frac{\partial \mathbf{F}}{\partial x}$, tells you how the vector field $\mathbf{F}$ is changing as you move a tiny bit in the positive $x$-direction.

*   **$\frac{\partial P}{\partial x}$:** Measures how the **$x$-component** of the vector is changing as you move in the $x$-direction.
*   **$\frac{\partial Q}{\partial x}$:** Measures how the **$y$-component** of the vector is changing as you move in the $x$-direction.

In essence, the partial derivative of a vector field is a new vector field that describes the **rate of change of the original vectors** as you move along a specific axis. It's a fundamental step used in more advanced concepts like **Divergence** and **Curl**! 

## Plot Vector Field

```python
X = np.arange(-5, 5, 0.025)
Y = np.arange(-5, 5, 0.025)
Xm, Ym = np.meshgrid(X, Y)
vf = lambda x, y : (np.multiply(x, y), x**2 - y**2)
U, V = vf(Xm, Ym)

plt.figure()
plt.title('Vector field f(X, Y)')
Clr = np.hypot(U, V)
d = 20
Q = plt.quiver(Xm[::d, ::d], Ym[::d, ::d], U[::d, ::d], V[::d, ::d],
               Clr[::d, ::d], pivot='mid', cmap=cm.coolwarm)
```

## Plot Derivative of Vector Field

```python
dvf_dx = lambda x, y : (y, 2 * x)
dvf_dy = lambda x, y : (x, -2 * y)
dU_dx, dV_dx = dvf_dx(Xm, Ym)
dU_dy, dV_dy = dvf_dy(Xm, Ym)

plt.figure()
plt.title('Vector field of df(X, Y) wrt x')
Clr = np.hypot(dU_dx, dV_dx)
d = 20
Q = plt.quiver(Xm[::d, ::d], Ym[::d, ::d], dU_dx[::d, ::d], dV_dx[::d, ::d],
               Clr[::d, ::d], pivot='mid', cmap=cm.coolwarm)

plt.figure()
plt.title('Vector field of df(X, Y) wrt y')
Clr = np.hypot(dU_dy, dV_dy)
d = 20
Q = plt.quiver(Xm[::d, ::d], Ym[::d, ::d], dU_dy[::d, ::d], dV_dy[::d, ::d],
               Clr[::d, ::d], pivot='mid', cmap=cm.coolwarm)
```
